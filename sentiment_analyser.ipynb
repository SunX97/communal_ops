```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased")
```

```python
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
sentence = "I love this movie!"
result = sentiment_analyzer(sentence)
print(result)
```

```python
dataset =[
          "This movie is fantastice!",
          "I didn't enjoy the food at all",
          "The concert was absolutely amazing",
          "The service was slow and disappointing",
          "I love the new phone's features!"
]
# Apply sentiment analysis to the entire dataset
results = sentiment_analyzer(dataset)

# Iterate through sentences and their corresponding results
for sentence, result in zip(dataset, results):
  print(f"Sentence: {sentence}")
  print(f"Sentiment: {result['label']} (Confidence: {result['score']:.4f})\n")
```
