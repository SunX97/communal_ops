```python
import time
import sys
from transformers import pipeline

def typing_effect(text):
    """Creates a typing effect for the given text."""
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(0.05)
    print() # for a new line after the text

def main():
    """
    The main function to run the translation chatbot.
    """
    # Load the pre-trained English-to-French translation model
    try:
        translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
    except Exception as e:
        print(f"Error loading the model: {e}")
        return

    print("English-to-French Translation Chatbot")
    print("Enter 'quit' to exit.")
    print("-" * 30)

    while True:
        english_text = input("You: ")

        if english_text.lower() == 'quit':
            print("Chatbot: Au revoir!")
            break

        try:
            # Translate the text
            translation = translator(english_text, max_length=512)
            french_text = translation[0]['translation_text']

            # Display the translation with a typing effect
            sys.stdout.write("Chatbot: ")
            sys.stdout.flush()
            typing_effect(french_text)

        except Exception as e:
            print(f"An error occurred during translation: {e}")

if __name__ == "__main__":
    main()
```
